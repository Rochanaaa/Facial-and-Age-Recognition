```
import matplotlib.pyplot as plt
import numpy as np

base_path = "drive/My Drive/CE4042 : Neural Network & Deep Learning/Project/"
dataset_prefix = "dataset/"
results_prefix = "results/"
plots_prefix = "plots/"
```

```
data = np.genfromtxt(base_path + results_prefix + 'efficient_final_unfreezed.csv', delimiter=',')
data = data[1:]
print(data, len(data))
print(data[1][1])

train_acc = []
train_loss = []
val_acc = []
val_loss = []
for i in range(len(data)):
    train_acc.append(data[i][1])
    train_loss.append(data[i][2])
    val_acc.append(data[i][3])
    val_loss.append(data[i][4])

# plot
plt.figure(1, figsize=(12, 7))
plt.plot(range(1, len(train_acc) + 1), train_acc, label='Train')
plt.plot(range(1, len(val_acc) + 1), val_acc, label='Test')
plt.title('Train-Test accuracy vs Epochs (Transfer Learning)')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(loc='lower right')
plt.savefig(base_path + plots_prefix + 'acc.png')
plt.show()

plt.figure(2, figsize=(12, 7))
plt.plot(range(1, len(train_loss) + 1), train_loss, label='Train')
plt.plot(range(1, len(val_loss) + 1), val_loss, label='Test')
plt.title('Train-Val loss vs Epochs (Transfer Learning)')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(loc='upper right')
plt.savefig(base_path + plots_prefix + 'loss.png')
plt.show()
```
